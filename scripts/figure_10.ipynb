```python
import pandas as pd
import numpy as np
import os
from glob import glob
import pickle as pickle

from collections import Counter
```

### Change PATH variable to match your own local file path:

```python
PATH = "/change-file-path-to/reproducibility/evaluation"
EXT = "*.csv"
all_csv_files = [file
                 for path, subdir, files in os.walk(PATH)
                 for file in glob(os.path.join(path, EXT))]

all_csv_files = [e for e in all_csv_files]

# print(all_csv_files)
```

```python
dfs = pd.DataFrame()
for f in all_csv_files:
    df = pd.read_csv(f, on_bad_lines='skip')
    display(df)
    dfs = dfs.append(df)
```

```python
dfs = dfs[['Conference', 'Year', 'Paper Title', 'Available', 'Compiles', 'Artifact Evaluation', 'Link for Code', 'Funding']]
dfs = dfs.rename(columns={'Title': 'Paper Title'})
dfs
```

```python
dfs = dfs.dropna(subset=['Paper Title'])
dfs
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
newStopWords = ['grant','supported', 'by', 'based', 'on', 'this', 'work', 'was', 'supported', 'partially', 
                'part', 'specified', 'no.', 'no', 'number', 'none', 'specified', 'specificed', 'acquisition', 
                'gift', 'funded', 'sponsored', 'grants', 'gratefully', 'acknowledge', 'cofunded', 'not', 'none']
stop.extend(newStopWords)

dfs.dropna(subset=['Compiles'], inplace=True)

dfs['Compiles'] = dfs['Compiles'].astype(str)
dfs['Compiles'] = dfs['Compiles'].str.split(';').str[0]
dfs['Compiles'] = dfs['Compiles'].str.split(',').str[0]

dfs['Compiles'] = dfs['Compiles'].astype(float)
dfs['Compiles'] = dfs['Compiles'].astype(int)
dfs['Compiles'] = pd.DataFrame(np.where((dfs['Compiles'] == 0) | (dfs['Compiles'] == 1),
                                                 "Compiles", "Does Not Compile"))

dfs.dropna(subset=['Funding'], inplace=True)

dfs['Funding'] = dfs['Funding'].str.replace(r"\(.*?\)", "", regex=True)
dfs['Funding'] = dfs['Funding'].str.split(';').str[0]
dfs['Funding'] = dfs['Funding'].str.split(',').str[0]
dfs['Funding'] = dfs['Funding'].str.split('/').str[0]


dfs['Funding'] = dfs['Funding'].str.replace('\d+', '')


dfs['Funding'] = dfs['Funding'].astype(str)
# funding_dfs['Funding'] = funding_dfs['Funding'].str.replace(r"\(.*\)","")

dfs['Funding'] = dfs['Funding'].replace('- ', '', regex=True)
dfs['Funding'] = dfs['Funding'].replace('-', '', regex=True)
dfs['Funding'] = dfs['Funding'].replace('.', '', regex=False)
dfs['Funding'] = dfs['Funding'].replace('(  ', '', regex=False)

dfs['Available'] = dfs['Available'] = dfs['Available'].str.split(',').str[0]
dfs['Available'] = dfs['Available'] = dfs['Available'].str.split(':').str[0]
dfs['Available'] = dfs['Available'] = dfs['Available'].str.split('(').str[0]
dfs['Available'] = dfs['Available'].replace('Research Group Website', '2', regex=False)
dfs['Available'] = dfs['Available'].replace('Google Drive', '2', regex=False)

dfs['Funding'] = dfs['Funding'].str.strip()

dfs['Funding'] = dfs['Funding'].str.lower()

dfs['Funding'] = dfs['Funding'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

dfs['Funding'] = dfs['Funding'].replace({'national science foundation':'nsf', 'ational':'national'})
dfs['Funding'] = dfs['Funding'].replace({'nsfc':'national natural science foundation china'})
dfs['Funding'] = dfs['Funding'].replace({'ministry education':'ministry of education, singapore'})


dfs['Funding'] = dfs['Funding'].replace({'office naval research':'onr'})
dfs['Funding'] = dfs['Funding'].replace({'onr n':'onr'})

dfs['Funding'] = dfs['Funding'].replace({'german research foundation':'dfg'})
dfs['Funding'] = dfs['Funding'].replace({'deutsche forschungsgemeinschaft (dfg':'dfg'})
dfs['Funding'] = dfs['Funding'].replace({'deutsche forschungsgemeinschaft (dfg)':'dfg'})

dfs['Funding'] = dfs['Funding'].str.replace('national science foundation','nsf')

dfs.loc[dfs['Funding'].str.contains('defense advanced research projects agency'), 'Funding'] = 'darpa'
dfs.loc[dfs['Funding'].str.contains('national key research development program china'), 'Funding'] = 'national key r&d program china'
dfs.loc[dfs['Funding'].str.contains('nprps'), 'Funding'] = 'qnrf qatar national research fund'
dfs.loc[dfs['Funding'].str.contains('samsung'), 'Funding'] = 'samsung'
dfs.loc[dfs['Funding'].str.contains('huawei'), 'Funding'] = 'huawei'
dfs.loc[dfs['Funding'].str.contains('national science found'), 'Funding'] = 'nsf'
dfs.loc[dfs['Funding'].str.contains('european union'), 'Funding'] = 'european union'
dfs.loc[dfs['Funding'].str.contains('national science found ation'), 'Funding'] = 'nsf'
dfs.loc[dfs['Funding'].str.contains('nsf'), 'Funding'] = 'nsf'
dfs.loc[dfs['Funding'].str.contains('epsrc'), 'Funding'] = 'epsrc'
dfs.loc[dfs['Funding'].str.contains('national key r&d'), 'Funding'] = 'national key r&d'
dfs.loc[dfs['Funding'].str.contains('u.s. army'), 'Funding'] = 'u.s. army'
dfs.loc[dfs['Funding'].str.contains('darpa'), 'Funding'] = 'darpa'
dfs.loc[dfs['Funding'].str.contains('eu'), 'Funding'] = 'european union'
dfs.loc[dfs['Funding'].str.contains('german research foundation'), 'Funding'] = 'german research foundation'
dfs.loc[dfs['Funding'].str.contains('german federal ministry education'), 'Funding'] = 'german federal ministry education research'
dfs.loc[dfs['Funding'].str.contains('german ministry education'), 'Funding'] = 'german federal ministry education research'
dfs.loc[dfs['Funding'].str.contains('google'), 'Funding'] = 'google'
dfs.loc[dfs['Funding'].str.contains('information & communications technology'), 'Funding'] = 'korean government'
dfs.loc[dfs['Funding'].str.contains('australian government'), 'Funding'] = 'australian government'
dfs.loc[dfs['Funding'].str.contains('u.s. office naval research'), 'Funding'] = 'onr'
dfs.loc[dfs['Funding'].str.contains('office naval research'), 'Funding'] = 'onr'
dfs.loc[dfs['Funding'].str.contains('msit'), 'Funding'] = 'korea government'
dfs.loc[dfs['Funding'].str.contains('institute information & communica'), 'Funding'] = 'korea government'
dfs.loc[dfs['Funding'].str.contains('institute information & communica'), 'Funding'] = 'korea government'
dfs.loc[dfs['Funding'].str.contains('korea government'), 'Funding'] = 'korean government'
```

```python
group = dfs.groupby(['Funding', 'Compiles']).agg(compiles_count=('Compiles','count')).reset_index()
# group = group[(group.Compiles != "Does Not Compile")]
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
group = group.sort_values('compiles_count', ascending=True)
group['Funding'].replace('', np.nan, inplace=True)
group.dropna(subset=['Funding'], inplace=True)
# group = group.tail(20)
group
```

```python
group.loc[group['Funding'] == "nsf", ['Funding']] = 'NSF'
group.loc[group['Funding'] == "onr", ['Funding']] = 'Office of Naval Research'
# funding_group.loc[funding_group['Funding'] == "onr n", ['Funding']] = 'Office of Naval Research'
group.loc[group['Funding'] == "defense advanced resarch projects agency", ['Funding']] = 'DARPA'


group.loc[group['Funding'] == "ministry education", ['Funding']] = 'Ministry of Education'
group.loc[group['Funding'] == "research fund ku lueven", ['Funding']] = 'KU Leuven Research Fund'
# funding_group.loc[funding_group['Funding'] == "nsf cns-", ['Funding']] = 'NSF CNS'
group.loc[group['Funding'] == "national key r&d", ['Funding']] = 'National Key R&D (China)'
group.loc[group['Funding'] == "european research council", ['Funding']] = 'European Research Council'
group.loc[group['Funding'] == "national natural science foundation china", ['Funding']] = 'NSFC (China)'
group.loc[group['Funding'] == "dfg", ['Funding']] = 'DFG (Germany)'
group.loc[group['Funding'] == "darpa", ['Funding']] = 'DARPA'
group.loc[group['Funding'] == "nsf cns", ['Funding']] = 'NSF CNS'
group.loc[group['Funding'] == "nserc", ['Funding']] = 'NSERC (Canada)'
group.loc[group['Funding'] == "european union", ['Funding']] = 'European Union'
group.loc[group['Funding'] == "university georgia research foundation", ['Funding']] = 'University of Georgia'
group.loc[group['Funding'] == "u.s. department state", ['Funding']] = '/U.S. Department of State'
group.loc[group['Funding'] == "meta research", ['Funding']] = 'META'
group.loc[group['Funding'] == "national research foundation", ['Funding']] = 'National Research Foundation'
group.loc[group['Funding'] == "epsrc", ['Funding']] = 'EPSRC'
group.loc[group['Funding'] == "purdue university", ['Funding']] = 'Purdue University'
group.loc[group['Funding'] == "deutsche forschungsgemeinschaft", ['Funding']] = 'DFG (Germany)'
group.loc[group['Funding'] == "german federal ministry education research", ['Funding']] = 'BMBF (Germany)'
group.loc[group['Funding'] == "korea government", ['Funding']] = 'Korean Government'
group.loc[group['Funding'] == "korean government", ['Funding']] = 'Korean Government'
group.loc[group['Funding'] == "academy finland", ['Funding']] = 'Academy of Finland'
group.loc[group['Funding'] == "google", ['Funding']] = 'Google'
```

```python
# group = group.sort_values('compiles_count', ascending=True)
group = group.tail(20).sort_values(by='compiles_count', ascending=True)

group = group.pivot_table(index=["Funding"], columns='Compiles', values='compiles_count', aggfunc='sum')

group = group.reindex(group['Compiles'].sort_values(ascending=False).index)

graph_group = group.sort_values('Does Not Compile', ascending=False)
ax = graph_group.plot(kind='barh', stacked=True)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_color('#DDDDDD')

ax.legend(['Compiles Count'])

ax.set_axisbelow(True)
ax.xaxis.grid(False)
ax.yaxis.grid(True, color='#EEEEEE')

new_ticks = np.linspace(0, 90, 11)  
ax.set_xticks(new_ticks)

labels = ['Compiles', 'Does Not Compile']
ax.legend(labels=labels)

fig = ax.get_figure()
# fig.savefig("comp_by_agency.png",dpi=1200,bbox_inches = "tight")
```
